```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
# removing all null vales in 'RespondentID' column
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
star_wars.columns
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

No null values in this column

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

350 null values present

```python
# converting/mapping above columns into True/False values

yes_no = {
    'Yes': True,
    'No': False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)

star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

star_wars.head()
```

```python
# converting episode names for each respondent to True/False values

movies_watched = {
    'Star Wars: Episode I  The Phantom Menace':True,
    'Star Wars: Episode II  Attack of the Clones':True,
    'Star Wars: Episode III  Revenge of the Sith':True,
    'Star Wars: Episode IV  A New Hope':True,
    'Star Wars: Episode V The Empire Strikes Back':True,
    'Star Wars: Episode VI Return of the Jedi':True,
    np.NaN:False,
}

for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(movies_watched)
```

```python
# renaming column names. 'seen_1' for 'Star Wars: Episode I The Phantom Menace' and so on

star_wars = star_wars.rename(columns={
    'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
    'Unnamed: 4':'seen_2',
    'Unnamed: 5':'seen_3',
    'Unnamed: 6':'seen_4',
    'Unnamed: 7':'seen_5',
    'Unnamed: 8':'seen_6'
})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# renaming columns which display ranking of each movie by a respondent

star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
    'Unnamed: 10':'ranking_2',
    'Unnamed: 11':'ranking_3',
    'Unnamed: 12':'ranking_4',
    'Unnamed: 13':'ranking_5',
    'Unnamed: 14':'ranking_6'
})

star_wars.head()
```

```python
# let's figure out the highest ranked movie (lower ranking is better)

%matplotlib inline
import matplotlib.pyplot as plt

bar = star_wars[star_wars.columns[9:15]].mean()

plt.bar(np.arange(6),bar, align='center')
plt.xticks(np.arange(6),('Episode I','Episode II','Episode III','Episode IV','Episode V', 'Episode VI'),rotation=30)
plt.ylabel('Average rating (Lower is better)')
```

## Average Rankings

So far, we have cleaned the data, renamed a few columns and calculated the mean value of movie rankings.

It looks like people still prefer the original Star Wars movies. The most preferred movie was Episode V - 'The Empire Strikes Back' which was released in 1980.

Two important things to note here:
1. Lower ranking is a better rank
2. Episode I,II and III are the most recent movies. Episode IV, V and VI are older.

```python
# let's find out the Most Viewed Movie
total_viewers = star_wars[star_wars.columns[3:9]].sum()

plt.bar(np.arange(6),total_viewers,align='center')
plt.xticks(np.arange(6),('Episode I','Episode II','Episode III','Episode IV','Episode V', 'Episode VI'),rotation=30)
plt.ylabel('Number of Viewers')
```

## Number of viewers

Not surprises here. The most viewed movie was Episode V, followed up Episode VI. This confirms our thoughts earlier as Episode V being the most preferred movie. These results correlate strongly with our ranking results.
